```python
X = df.iloc[:,:4].values
X
```

```python
y = df['variety'].values
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 82)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
len(X_train)
```

# using  linear kernel

```python
from sklearn.svm import SVC
svcclassifier = SVC(kernel = 'linear', random_state = 0,gamma)
svcclassifier.fit(X_train, y_train)
```

```python
y_pred = svcclassifier.predict(X_test)
print(y_pred)
```

```python
a=svcclassifier.score(X_test,y_test)
a
```

# using ploynomial kernel

```python
from sklearn.svm import SVC
svcclassifier = SVC(kernel = 'poly', random_state = 0)
svcclassifier.fit(X_train, y_train)
```

```python
y_pred = svcclassifier.predict(X_test)
print(y_pred)
```

```python
b=svcclassifier.score(X_test,y_test)
b
```

# using rbf kernel

```python
from sklearn.svm import SVC
svcclassifier = SVC(kernel = 'rbf', random_state = 0)
svcclassifier.fit(X_train, y_train)
```

```python
y_pred = svcclassifier.predict(X_test)
print(y_pred)
```

```python
c=svcclassifier.score(X_test,y_test)
c
```

# using sigmoid function

```python
from sklearn.svm import SVC
svcclassifier = SVC(kernel = 'sigmoid', random_state = 0)
svcclassifier.fit(X_train, y_train)
```

```python
y_pred = svcclassifier.predict(X_test)
print(y_pred)
```

```python
d=svcclassifier.score(X_test,y_test)
d
```
